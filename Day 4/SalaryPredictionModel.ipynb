```python
import numpy as np

# 1. Data
X = np.array([1, 2, 3, 4, 5])  # Years of experience
Y = np.array([40000, 50000, 60000, 70000, 80000])  # Salary

# 2. Initialize weights and bias
w = np.random.randn()  # random weight
b = np.random.randn()  # random bias

# 3. Learning rate
lr = 0.01

# 4. Training loop
for epoch in range(1000):
    # Forward pass: prediction = dot product + bias
    Y_pred = X * w + b   # dot product for each x_i

    # Loss = Mean Squared Error
    loss = np.mean((Y - Y_pred)**2)

    # Backward pass: gradients
    dw = -2 * np.mean(X * (Y - Y_pred))  # derivative w.r.t weight
    db = -2 * np.mean(Y - Y_pred)         # derivative w.r.t bias

    # Update weights
    w -= lr * dw
    b -= lr * db

    if epoch % 100 == 0:
        print(f'Epoch {epoch}, Loss: {loss:.2f}, w: {w:.2f}, b: {b:.2f}')

# 5. Final model
print(f'Final model: salary = {w:.2f} * years + {b:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Data
X = np.array([1, 2, 3, 4, 5])  # Years of experience
Y = np.array([40000, 50000, 60000, 70000, 80000])  # Salary

# 2. Initialize weights and bias
w = np.random.randn()
b = np.random.randn()

# 3. Learning rate
lr = 0.01

# Store values for plotting
losses = []
weights = []
biases = []
predictions_over_time = []

# 4. Training loop
for epoch in range(1000):
    Y_pred = X * w + b
    loss = np.mean((Y - Y_pred)**2)

    dw = -2 * np.mean(X * (Y - Y_pred))
    db = -2 * np.mean(Y - Y_pred)

    w -= lr * dw
    b -= lr * db

    # Store for plotting
    losses.append(loss)
    weights.append(w)
    biases.append(b)
    if epoch % 100 == 0 or epoch == 999:
        predictions_over_time.append((epoch, Y_pred.copy()))

# Final model
final_w = w
final_b = b

# Plot 1: Data and Final Regression Line
plt.figure(figsize=(16, 12))

plt.subplot(2, 2, 1)
plt.scatter(X, Y, label='Actual Data')
plt.plot(X, X * final_w + final_b, color='red', label='Learned Line')
plt.title('Actual Data vs Learned Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()

# Plot 2: Loss Over Epochs
plt.subplot(2, 2, 2)
plt.plot(losses)
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')

# Plot 3: Weight and Bias over Epochs
plt.subplot(2, 2, 3)
plt.plot(weights, label='Weight (w)')
plt.plot(biases, label='Bias (b)')
plt.title('Weight and Bias Evolution')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.legend()

# Plot 4: Predictions over time
plt.subplot(2, 2, 4)
for epoch, preds in predictions_over_time:
    plt.plot(X, preds, label=f'Epoch {epoch}')
plt.scatter(X, Y, color='black', label='Actual')
plt.title('Predictions Over Time')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()

plt.tight_layout()
plt.show()
```
